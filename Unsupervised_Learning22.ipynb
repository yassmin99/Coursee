```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import mlxtend
```

```python
dataset = [['Skirt', 'Sneakers', 'Scarf', 'Pants', 'Hat'],

    ['Sunglasses', 'Skirt', 'Sneakers', 'Pants', 'Hat'],

    ['Dress', 'Sandals', 'Scarf', 'Pants', 'Heels'],

    ['Dress', 'Necklace', 'Earrings', 'Scarf', 'Hat', 'Heels', 'Hat'],

   ['Earrings', 'Skirt', 'Skirt', 'Scarf', 'Shirt', 'Pants']]
```

```python
from mlxtend.preprocessing import TransactionEncoder
te=TransactionEncoder()
te_ary=te.fit(dataset).transform(dataset)    #Apply one-hot-encoding on our dataset
df=pd.DataFrame(te_ary, columns=te.columns_)  #Creating a new DataFrame from our Numpy array
df
```

```python
from mlxtend.frequent_patterns import apriori
apriori(df, min_support=0.6)
```

```python
frequent_itemsets=apriori(df, min_support=0.6, use_colnames=True)
frequent_itemsets
```

```python
from mlxtend.frequent_patterns import association_rules 
association_rules(frequent_itemsets,metric="confidence",min_threshold=0.7)
```

```python
from mlxtend.frequent_patterns import association_rules 
association_rules(frequent_itemsets,metric="lift",min_threshold=1.25)
```

```python
list_values = df1.values.tolist()
cleaned_tx_lol = []
for tx_list in list_values:
    cleaned_tx_list = []
    for tx in tx_list:
        if type(tx) != str:
            continue
        else:
            cleaned_tx_list.append(tx)
    cleaned_tx_lol.append(cleaned_tx_list)
  
```

```python

df1.head()
```

```python
from mlxtend.preprocessing import TransactionEncoder
te=TransactionEncoder()
te_ary=te.fit(cleaned_tx_lol).transform(cleaned_tx_lol)   
dff=pd.DataFrame(te_ary, columns=te.columns_)  
dff
```

```python
apriori(dff, min_support=0.01)
```

```python
frequent_itemsets1=apriori(dff, min_support=0.05, use_colnames=True)
frequent_itemsets1
```

```python
association_rules(frequent_itemsets1,metric="confidence",min_threshold=0.01)
```

```python
association_rules(frequent_itemsets1,metric="lift",min_threshold=1.0000001)
frequent_itemsets1.max()
```
